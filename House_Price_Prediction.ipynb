**importing all the necessery libaries**

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

**loading the data set**

```python
df1 = pd.read_csv("/content/housing.csv")
```

**checking for null values in data**

```python
df1.info()
```

```python
df1.isnull().sum()
```

```python
df = df1.dropna()
df.info()
X = df.drop(['ocean_proximity'],axis=1)
```

```python
Y= df['ocean_proximity']
Z = pd.get_dummies(df['ocean_proximity'])
data_set = X.join(Z)
data_set
```

```python
X = data_set.drop(['median_house_value'],axis=1)
Y = data_set['median_house_value']
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state = 2)
```

```python
data_set.hist(figsize =(15,8))
```

```python
train_data_set = X_train.join(Y_train)
plt.figure(figsize=(15,8))
sns.heatmap(train_data_set.corr(), annot=True, cmap = 'YlGnBu')
```

```python
plt.figure(figsize =(15,8))
sns.scatterplot( x='latitude', y = 'longitude', data = train_data_set, hue = 'median_house_value', palette = 'coolwarm')
```

```python
plt.figure(figsize = (15,8))
train_data_set['bedrroom_ratio'] = train_data_set['total_bedrooms']/train_data_set['total_rooms']
sns.heatmap(train_data_set.corr(), annot = True, cmap = 'YlGnBu')
```

```python
x_train = train_data_set.drop(['median_house_value'], axis=1)
y_train = train_data_set['median_house_value']
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
model.score(x_train, y_train)
```

```python
model.fit(x_train, y_train)
y_train_predict = model.predict(x_train)
model.score(x_train, y_train)
```

```python
test_data_set = X_test.join(Y_test)
test_data_set['bedroom_ratio'] = test_data_set['total_bedrooms']/test_data_set['total_rooms']
X_test = test_data_set.drop(['median_house_value'],axis=1)
Y_test = test_data_set['median_house_value']
model.fit(X_test, Y_test)
model.score(X_test, Y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(x_train, y_train)
model.score(x_train, y_train)
```
